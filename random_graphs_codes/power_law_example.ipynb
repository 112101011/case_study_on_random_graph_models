```python
import matplotlib.pyplot as plt

# Let us see one simple example of power law
# Let a be th area of the square
# and s be the side of the square

s_values = [i for i in range(0, 100000)]
a_values = [s*s for s in s_values]

plt.plot(s_values, a_values, label = "a = s\u00B2" )
plt.legend()
plt.ylabel("Area a -> ")
plt.xlabel("Side of a sqaure s->")
plt.title("Area of the square vs side of a square")
plt.show()
```
