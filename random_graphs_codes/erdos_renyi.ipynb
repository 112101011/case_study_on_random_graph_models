```python
import random
# notice the diversity in p_values
p_values = [0.01, 0.3, 0.8]

# number of vertices values
n_values = [ 50, 1000, 1500]

for p in p_values:
    for n in n_values:
        # G(n, p) model
        
        adj = dict()
        for i in range(1, n+1):
            adj[i] = list()
    
        for u in range(1, n+1):
            for v in range(u+1, n+1):
                if random.uniform(0, 1) < p:
                    # add edge
                    adj[u].append(v)
                    adj[v].append(u)

        print("G(n = ", n, ", p = ", p)
        erdos_renyi_model_checker.check(adj)
    
```

Binomial Distribution: The degree distribution of an ER model follows a binomial distribution Bin(n−1,p), which approximates a Poisson distribution Poisson(λ) for large n, where λ=(n−1)p.
