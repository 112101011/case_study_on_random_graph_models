```python
# creating an object for analyzing facebook graph dataset
barabasi_albert_model_checker = check_properties()
```

```python
import random
max_time = 1000
adj = dict()
adj[1] = [2, ]
adj[2] = [3, ]
adj[3] = [1, ]
# m is the number of edges added when new vertex is introduced
m = 4 
next_vertex = 4
print("Time = ", 0)
barabasi_albert_model_checker.check(adj)
times = dict()
times[10] = True
times[50] = True
times[100] = True
times[500] = True
times[1000] = True

for t in range(1, max_time + 1):
    prob = []
    vertices = list(adj.keys())
    total_degree = 0
    for u in vertices:
        total_degree += len(adj[u])
    for u in vertices:
        prob.append(len(adj[u])/total_degree)
    # we are considering parallel edges, generally they are considered in this model
    neighbours = random.choices(vertices, weights = prob, k = m)
    # print(neighbours)
    for u in neighbours:
        adj[u].append(next_vertex)
        if adj.get(next_vertex) == None:
            adj[next_vertex] = [u, ]
        else:
            adj[next_vertex].append(u)
    next_vertex += 1
    if times.get(t) != None:
        print("Time = ", t)
        barabasi_albert_model_checker.check(adj)
    
    
```
