```python
f = open("../face_book_dataset/facebook_combined.txt")
adj = dict()
for line in f:
    li = list()
    for ele in line.split():
        li.append(int(ele))
    u = li[0]
    v = li[1]
    if(adj.get(u) == None):
        adj[u] = [v, ]
    else:
        adj[u].append(v)

    if(adj.get(v) == None):
        adj[v] = [u, ]
    else:
        adj[v].append(u)
```
